## Helper Functions

```python
def findRating(stringInput):
    output = ''
    i = 0
    for j in stringInput:
        if (j.isdigit() or j == '.' or j == '/') and i < 50:
            output += str(j)
        i += 1
    while output.find('.') != -1 and output[-1] == '.':
        output = output[:-1]
    while output.find('.') != -1 and output[0] == '.':
        output = output[1:]
    if output == '':
        return '**No Rating Found**'
    return output


def seperateComments(commentList, nameList, commentString):
    startList = []
    i = 0
    for name in nameList:
        tempString = commentString.lower()
        tempName1 = name.lower()
        tempName2 = name.lower().split()[0]
        tempName3 = name.lower().split()[0][:3]
        try: commentList[i]
        except: commentList.append([])
        i += 1
        if tempString.find(tempName1) != -1:
            startList.append(tempString.find(tempName1))
            tempString = tempString[tempString.find(tempName1) + len(tempName1):]
        elif tempString.find(tempName2) != -1:
            startList.append(tempString.find(tempName2))
            tempString = tempString[tempString.find(tempName2) + len(tempName2):]
        elif tempString.find(tempName3) != -1:
            startList.append(tempString.find(tempName3))
            tempString = tempString[tempString.find(tempName3) + len(tempName3):]
        else:
            startList.append(-1)

    sortList = startList.copy()
    sortList.sort()
    j = 0
    for i in startList:
        if i == -1:
            commentList[j].append(nameList[j] + ': **No Comment Found.** from person' + '\n')
        elif i == max(startList):
            commentList[j].append(commentString[i:] + '\n')
            pass
        elif j < len(startList) - 1:
            commentList[j].append(commentString[i:sortList[sortList.index(i)+1]] + '\n')
        elif j == len(startList) - 1:
            commentList[j].append(commentString[i:sortList[sortList.index(i)+1]] + '\n')
        j += 1

    return commentList
```

## Testing Taking Input

```python
# Copy and paste entire comments from one person into something maybe a text file
tempList = []
for group in groups:
    i = 0
    print(f'group {group}')
    for person in group:
        stringInput = input("Enter comments from person {}: ".format(i))
        if stringInput == 'back':
            if i != 0:
                i -= 1
        elif stringInput == 'quit' or stringInput == 'next':
            break
        else:
            tempList += seperateComments(tempList, *group, stringInput)
            i += 1
    if stringInput == 'quit':
        break
```

## Making lists for simplification

```python
totalNumList = []
totalSumList = ['']
groupSumList = ['']

for i in range(0, 10):
    tempList = []
    for j in range(i):
        tempList.append(1)
    totalNumList.append(tempList)

for i in range(0, 9):
    tempString = ''
    for j in range(i):
        tempString += '25, '
    tempString += '25'
    totalSumList.append(tempString)

for i in range(0, 9):
    tempString = ''
    for j in range(i):
        tempString += '75, '
    tempString += '75'
    groupSumList.append(tempString)


print(totalNumList)
print(totalSumList)
print(groupSumList)
```

## Create Markdown Template

```python
output = ''
# totalNumList created above
# totalSumList created above
# groupSumList created above

for i in range(0, len(groups)):
    if i == 0:
        groupNum = i + 1
    else:
        groupNum = i + 2
    output += '# **Group {}**\n'.format(groupNum)
    for j in groups[i]:
        output += '## {}\n'.format(j)
        for k in groups[i]:
            output += '\t- {0} to {1}: \n'.format(k, j)
        output += '### {0} Total:\n> Score: {1}  \n> Grade: Median of [{2}] = **25**\n'.format(j, totalNumList[len(groups[i])], totalSumList[len(groups[i])])
    output += '## Group {} Score: \n'.format(groupNum)
    for m in groups[i]:
        output += '\t- {0}: \n'.format(m)
    output += '\t- Group {0} Total: \n> Score: {1} \n> Grade: Median of [{2}] = **75**\n'.format(groupNum, totalNumList[len(groups[i])], groupSumList[len(groups[i])], len(groups[i]))
    output += '\n* * *\n\n'
print(output)
```

## Append it to a file for use

```python
f = open("test_grades.md", "a")
f.write(output)
f.close()
```
